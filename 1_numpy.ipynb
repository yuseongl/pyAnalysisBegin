### 11. 배열 연결 (병합) : stack, hstack, vstack, dstack, tile

```python
##### hstack, vstack

arr1 = np.ones((2,3))
print(arr1)
arr2 = np.ones((1,3))
print(arr2)

#arr3 = np.hstack([arr1,arr2])
#print(arr3)

arr3 = np.vstack([arr1,arr2])
print(arr3)
```

```python
##### dstack : 차원의 증가

arr1 = np.ones((3,4))
arr2 = np.zeros((3,4))

print(arr1)
print(arr2)

arr3 = np.dstack([arr1,arr2])
print(arr3)

print(arr3.shape)
```

```python
##### stack : 차원의 증가(dstack을 단순화 시킨 작업)

arr1 = np.ones((3,4))
arr2 = np.zeros((3,4))

arr3 = np.stack([arr1,arr2])
print(arr3)
print(arr3.shape)

print('------------------------------')

arr4 = np.stack([arr1,arr2], axis=0)
print(arr4)

print('------------------------------')
arr4 = np.stack([arr1,arr2], axis=1)
print(arr4)
```

```python
##### tile

arr = np.array([0,1,2],[3,4,5])

print(np.tile(arr,2))

print('------------------------------')
```

### 18. 정수형 데이터 카운팅 : unique(), bincount()

```python
arr = np.array(['a','b','b','c','a'])

print(len(arr))
print(np.unique(arr))

print('---------------------------------')
idx,cnt = np.unique(arr, return_counts=True)
print(idx,cnt)
```

```python
# bincount : 발생하지 않은 사건도 카운팅

arr = np.array([1,1,2,2,2,3])

print(np.unique(arr,return_counts = True))
print(np.bincount(arr,minlength=6))
```

### 실습 : 영화 평점 데이터

+ https://grouplens.org/datasets/movielens/

```python
import numpy as np
data = np.loadtxt("data/ml-1m/ml-1m/ratings.dat",delimiter='::',dtype=np.int64)
```

```python
print(data)
print(data.shape)
```

```python
# 1행부터 20행까지 출력
import numpy as np
cnt = 1
for i in range(20):
    
    print('{}번'.format(cnt),data[i])
    cnt += 1
```

```python
# 21행 부터 30행까지 userid와 movieid만 출력

#1
print(data[20:30,:2])

# 2
cnt = 1
for i in range(10):
    #print('{}번'.format(cnt))
    print('{}번'.format(cnt),data[20+i:21+i,:2])
    cnt += 1
```

```python
# 전체 평점에 대한 평균

print(data[:,2].mean())
print(np.mean(data[:,2]))
```

```python
# 사용자별로 평점 평균(10명만 출력)

#1
avr = 0
num = 0
for i in range(1,11):
    num += 1
    avr = np.mean(data[data[:,0]==i][:,2])
    print('{}번 고객의 평균은'.format(num),avr)
    
    
#2
User_ID = np.unique(data[:,0])
mean_Rating =[]

for i in User_ID:
    data_user = data[data[:,0] == i,:]
    Rating = np.mean(data_user[:,2])
    mean_Rating.append([i, Rating])

for j in range(10):
    print("{}번 사용자의 평점 평균은 {}점".format(j+1, round(mean_Rating[j][1], 2)))
  
    
#3
a = np.unique(data[:,0])

for i in a:
    b = data[data[:,0]==i][:,2]
    avg = np.mean(b)
    
    print('{}번 사용자의 평점 평균은 {}점'.format(i, round(avg, 2)))
```

```python
# 위의 결과 중 최고 평점 평균 구하기
arrmean_m = np.array(mean_Rating)

print((np.argmax(arrmean_m[:, 1])+1),round(np.max(arrmean_m[:, 1]), 2))

maxscore = []
for i in np.arange(1, 11):
    score = np.mean(data[data[:, 0]==i][:, 2])
    maxscore.append(score)
np.max(maxscore)
```

```python
arrmean_m = np.array(mean_Rating)

print("평균의 최고값은 {}번의 {}점 입니다".format((np.argmax(arrmean_m[:, 1])+1),
                                             round(np.max(arrmean_m[:, 1]), 2)))
```

```python
score_lst = []
for i in np.arange(1, 11):
    score = np.mean(data[data[:, 0]==i][:, 2])
    score_lst.append(score)

np.max(score_lst)
```

```python
# 결과값 파일로 저장 
np.savetxt('data/ml-1m/id_rations.csv',arrmean_m,delimiter=',',fmt='%.1f')
```
