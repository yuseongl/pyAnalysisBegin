#### 5) 상세 페이지 데이터 가져오기

```python
url = df['ReadMore'][0]
html = requests.get(url, headers={'User-Agent':'Mozilla/5.0'})

soup = BeautifulSoup(html.text,'html.parser')
soup

detail = soup.find('p','addy').get_text()
print(detail)

detail_split = detail.split()
print(detail.split)

# 가격 가져오기
print(detail_split[0])
print(detail_split[0][:-1])

# 주소 가져오기
detail_split[1:-2]
' '.join(detail_split[1:-2])
```

```python
##### 각 데이터 수집 : 가격, 주소
##### jupyter notebook에서 상태 진행바를 보여주기
# conda install -c conda-forge tqdm

from tqdm import tqdm_notebook
from time import sleep

price = []
address = []

for n in tqdm_notebook(df.index):
    html = requests.get(df['ReadMore'][n], headers = {'USer-Agent':'Mozilla/5.0'})# 에러날경우 verify=False 추가
    soup = BeautifulSoup(html.text,'html.parser')
    
    detail = soup.find('p','addy').get_text()
    
    price.append(detail.split()[0][:-1])
    address.append(' '.join(detail_split()[1:-2]))
    
    #sleep(3)
```

```python
len(price),len(address)
print(price[:5])
print(address[:5])
```

```python
##### 두개의 정보를 DataFrame에 추가

df['Price']=price
df['Address']=address
df.head()
```
